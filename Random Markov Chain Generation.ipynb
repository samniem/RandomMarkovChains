# Random Markov Chains

```python
import numpy as np
from random import randint
import matplotlib.pyplot as plt
```

## Transition Matrix Generation
Change n value to change the number of states in the Markov chain. For example n = 4 would yield a Markov Chain such as the one shown below. Each transition from a state to a new state is assigned some probability. The possible transitions are indicated by arrows and the arrow strengths indicate the likelihood of each transition.


The initial starting state is generated randomly by changing one state value to 1 using randint function. The initial state is stored to array v. The transition matrix values are generated randomly using Numpy and each row is set to have a total probability of 1.

```python
#change n to generate a Markov chain of different size
n = 4

v = np.zeros(n,)
v[randint(0,3)] = 1
print("\nInitial starting vector is:\n", v, "\n")
T = np.random.rand(n,n)
for i in range(n):
    T[i] /= T[i].sum()
#If some value is 1 we will change it to 
print("The transition matrix T is\n", T)
```

# Future State Probabilities

To calculate the probability of being in each state after i iterations change the value of i to a desired number. Default i has been set to 10.

```python
#Change i to calculate a specific future state probability vector
i = 10

Ti = np.linalg.matrix_power(T,i)
print("The transition matrix values after i iterations:\n", Ti)
x=np.dot(v,Ti)
print("\nThe probability of being in each state after i =",i,"iterations:\n",x)
```

```python
plt.bar(np.arange(len(x)),x, width=0.9)
plt.xticks(np.arange(len(x)))
plt.ylabel("Probability of state")
plt.xlabel("States")
plt.show()
```
